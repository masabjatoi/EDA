## Feature Engeering

```python
train["SibSp"].value_counts()
```

```python
train[train["SibSp"] == 8]
```

```python
test[test["Ticket"] == "CA. 2343"]
```

```python
df = pd.concat([train , test])
```

```python
df
```

```python
df[df["Ticket"] == "CA. 2343"]
```

```python
df["Indivsual Fare"] = df["Fare"]/(df["SibSp"] + df["Parch"] + 1)
```

```python
df
```

```python
df["Family_size"]  = df["SibSp"] + df["Parch"] + 1
```

```python
## Family_Type

# 1 Alone
# 2-4 Small Family
# > 4 large family

def family(num):
  if num == 1:
    return "Alone"
  elif num > 1 and num <= 4:
    return "Small Family"
  else:
    return "Large Family"


df["Family_type"] = df["Family_size"].apply(family)
```

```python
df
```

```python
pd.crosstab(df["Survived"] ,df["Family_type"] , normalize = "columns" )* 100
```

```python
df["Surname"] = df["Name"].str.split(",").str.get(0)
```

```python
df
```

```python
df["Title"] = df["Name"].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0)
```

```python
df
```

```python
df["Title"].value_counts()
```

```python

df["Title"] = df["Title"].str.replace(r"Others\.\.*.*", "Others", regex=True)
df["Title"] = df["Title"].str.replace(r"Lady.", "Miss", regex=True)

value_counts = df["Title"].value_counts()
print(value_counts)
```

```python
pd.crosstab(df["Survived"] , df["Title"] , normalize= "columns") * 100
```

```python
df["Cabin"].isnull().sum()
```

```python
df["Cabin"].fillna("M" , inplace = True)
```

```python
df["Cabin"].value_counts()
```

```python
df["Deck"] = df["Cabin"].str[0]
```
